# Encode Categorical Columns

```python
# =========================
# Categorical Column Detection
# =========================

categorical_columns = X_df.select_dtypes(
    exclude=np.number
).columns.tolist()

print("Categorical Columns:")
print(categorical_columns)

# =========================
# Categorical Encoding
# =========================

ENCODING_METADATA = {}

for col in categorical_columns:

    unique_count = X_df[col].nunique()

    # -------------------------
    # One-Hot Encoding
    # -------------------------

    if unique_count <= 10:

        print(f"{col}: using ONE-HOT encoding")

        dummies = pd.get_dummies(
            X_df[col],
            prefix=col,
            drop_first=True
        )

        X_df = pd.concat(
            [X_df.drop(columns=[col]), dummies],
            axis=1
        )

        ENCODING_METADATA[col] = "onehot"

    # -------------------------
    # Label Encoding
    # -------------------------

    else:

        print(f"{col}: using LABEL encoding")

        encoder = LabelEncoder()

        X_df[col] = encoder.fit_transform(
            X_df[col].astype(str)
        )

        ENCODING_METADATA[col] = "label"

#rebuild numeric df
numeric_df = X_df.select_dtypes(
    include=np.number
)
```

```python
# =========================
# Scaling / Normalization
# =========================

# =========================
# Scaler Decision Engine
# =========================

scaler_scores = {

    "standard": 0,
    "minmax": 0,
    "robust": 0
}

# -------------------------
# Outlier Preference
# -------------------------

if (
    FEATURE_ENGINEERING_CONFIG[
        "prefer_robust_for_outliers"
    ]
    and
    DATASET_METADATA["outlier_fraction"]
    >
    FEATURE_ENGINEERING_CONFIG[
        "outlier_fraction_threshold"
    ]
):

    scaler_scores["robust"] += 2

    print("High outlier presence detected.")
    print("Increasing RobustScaler preference.\n")

# -------------------------
# Heavy Tail Preference
# -------------------------

if (
    DATASET_METADATA["average_kurtosis"]
    >
    FEATURE_ENGINEERING_CONFIG[
        "kurtosis_threshold"
    ]
):

    scaler_scores["robust"] += 3

    print("Heavy-tailed distributions detected.")
    print("Strong RobustScaler preference added.\n")

# -------------------------
# Skew Preference
# -------------------------

if (
    FEATURE_ENGINEERING_CONFIG[
        "prefer_minmax_for_skewed_data"
    ]
    and
    DATASET_METADATA["average_skew"]
    >
    FEATURE_ENGINEERING_CONFIG[
        "skew_threshold"
    ]
):

    scaler_scores["minmax"] += 1

    print("High skew detected.")
    print("Increasing MinMaxScaler preference.\n")

# -------------------------
# Sparse-like Preference
# -------------------------

if (
    FEATURE_ENGINEERING_CONFIG[
        "prefer_standard_for_sparse_data"
    ]
    and
    DATASET_METADATA["zero_fraction"]
    >
    FEATURE_ENGINEERING_CONFIG[
        "sparse_zero_threshold"
    ]
):

    scaler_scores["standard"] += 1

    print("Sparse-like dataset detected.")
    print("Increasing StandardScaler preference.\n")

# -------------------------
# Small Range Preference
# -------------------------

if (
    DATASET_METADATA["average_feature_range"]
    < 5
):

    scaler_scores["standard"] += 1

    print("Features already have relatively small ranges.")
    print("Increasing StandardScaler preference.\n")

# -------------------------
# Binary Feature Preference
# -------------------------

if (
    DATASET_METADATA["binary_fraction"]
    > 0.5
):

    scaler_scores["standard"] += 1

    print("Many binary-like features detected.")
    print("Increasing StandardScaler preference.\n")

# =========================
# Scaler Scores
# =========================

print("Scaler Scores:")
print(scaler_scores)
```

# Pre-Metadata

```python
# =========================
# Numeric Feature Statistics
# =========================

average_skew = (
    numeric_df.apply(skew).abs().mean()
)

average_kurtosis = (
    numeric_df.apply(kurtosis).mean()
)

zero_fraction = (
    (numeric_df == 0).mean().mean()
)

# -------------------------
# Outlier Statistics
# -------------------------

q1 = numeric_df.quantile(0.25)

q3 = numeric_df.quantile(0.75)

iqr = q3 - q1

outlier_mask = (

    (numeric_df < (q1 - 1.5 * iqr))
    |
    (numeric_df > (q3 + 1.5 * iqr))
)

outlier_fraction = (
    outlier_mask.mean().mean()
)

total_outliers = (
    outlier_mask.sum().sum()
)

# -------------------------
# Feature Range Statistics
# -------------------------

feature_ranges = (
    numeric_df.max() - numeric_df.min()
)

average_range = (
    feature_ranges.mean()
)

# -------------------------
# Binary Feature Statistics
# -------------------------

binary_fraction = np.mean([

    numeric_df[col].nunique() <= 2

    for col in numeric_df.columns
])

# -------------------------
# Missing Values
# -------------------------

numeric_missing_fraction = (
    numeric_df.isnull().mean().mean()
)

categorical_missing_fraction = (
    X_df.select_dtypes(
        exclude=np.number
    ).isnull().mean().mean()
)

# -------------------------
# Log Transform Statistics
# -------------------------

column_skewness = (
    numeric_df.apply(skew).abs()
)

high_skew_columns = column_skewness[
    column_skewness >
    FEATURE_ENGINEERING_CONFIG[
        "log_transform_skew_threshold"
    ]
]

positive_only_columns = [

    col

    for col in numeric_df.columns

    if (numeric_df[col] > 0).all()
]
```

# Missing Values Handling 

```python
# =========================
# Missing Value Decision Engine
# =========================

numeric_missing_scores = {

    "mean": 0,
    "median": 0,
    "drop": 0
}

categorical_missing_scores = {

    "mode": 0,
    "constant": 0,
    "drop": 0
}

# -------------------------
# High Missingness -> Drop
# -------------------------

if (
    FEATURE_ENGINEERING_CONFIG[
        "prefer_drop_for_high_missingness"
    ]
    and
    DATASET_METADATA[
        "numeric_missing_fraction"
    ]
    >
    FEATURE_ENGINEERING_CONFIG[
        "high_missingness_threshold"
    ]
):

    numeric_missing_scores["drop"] += 3

    print("High numeric missingness detected.")
    print("Increasing DROP preference.\n")

# -------------------------
# Outliers -> Median
# -------------------------

if (
    FEATURE_ENGINEERING_CONFIG[
        "prefer_median_for_outliers"
    ]
    and
    DATASET_METADATA[
        "outlier_fraction"
    ]
    >
    FEATURE_ENGINEERING_CONFIG[
        "numeric_outlier_threshold"
    ]
):

    numeric_missing_scores["median"] += 2

    print("Outliers detected.")
    print("Increasing MEDIAN preference.\n")

# -------------------------
# Low Outliers -> Mean
# -------------------------

if (
    FEATURE_ENGINEERING_CONFIG[
        "prefer_mean_for_normal_data"
    ]
    and
    DATASET_METADATA[
        "outlier_fraction"
    ]
    <=
    FEATURE_ENGINEERING_CONFIG[
        "numeric_outlier_threshold"
    ]
):

    numeric_missing_scores["mean"] += 1

    print("Numeric data appears relatively normal.")
    print("Increasing MEAN preference.\n")

# -------------------------
# High Categorical Missingness -> Drop
# -------------------------

if (
    DATASET_METADATA[
        "categorical_missing_fraction"
    ]
    >
    FEATURE_ENGINEERING_CONFIG[
        "high_missingness_threshold"
    ]
):

    categorical_missing_scores["drop"] += 2

    print("High categorical missingness detected.")
    print("Increasing DROP preference.\n")

# -------------------------
# Otherwise -> Mode
# -------------------------

else:

    categorical_missing_scores["mode"] += 1

    print("Categorical missingness appears manageable.")
    print("Increasing MODE preference.\n")

# =========================
# Missing Value Scores
# =========================

print("Numeric Missing Scores:")
print(numeric_missing_scores)

print("\nCategorical Missing Scores:")
print(categorical_missing_scores)
```

# Outlier Handling 

```python
# =========================
# Outlier Decision Engine
# =========================

outlier_strategy_scores = {

    "clip": 0,
    "remove": 0,
    "ignore": 0
}

# -------------------------
# Moderate Outliers -> Clip
# -------------------------

if (
    FEATURE_ENGINEERING_CONFIG[
        "prefer_clipping_for_moderate_outliers"
    ]
    and
    DATASET_METADATA[
        "outlier_fraction"
    ]
    >
    FEATURE_ENGINEERING_CONFIG[
        "moderate_outlier_threshold"
    ]
    and
    DATASET_METADATA[
        "outlier_fraction"
    ]
    <=
    FEATURE_ENGINEERING_CONFIG[
        "extreme_outlier_threshold"
    ]
):

    outlier_strategy_scores["clip"] += 2

    print("Moderate outlier presence detected.")
    print("Increasing CLIP preference.\n")

# -------------------------
# Extreme Outliers -> Remove
# -------------------------

if (
    FEATURE_ENGINEERING_CONFIG[
        "prefer_removal_for_extreme_outliers"
    ]
    and
    DATASET_METADATA[
        "outlier_fraction"
    ]
    >
    FEATURE_ENGINEERING_CONFIG[
        "extreme_outlier_threshold"
    ]
):

    outlier_strategy_scores["remove"] += 3

    print("Extreme outlier presence detected.")
    print("Increasing REMOVE preference.\n")

# -------------------------
# Low Outliers -> Ignore
# -------------------------

if (
    FEATURE_ENGINEERING_CONFIG[
        "prefer_ignore_for_low_outliers"
    ]
    and
    DATASET_METADATA[
        "outlier_fraction"
    ]
    <=
    FEATURE_ENGINEERING_CONFIG[
        "moderate_outlier_threshold"
    ]
):

    outlier_strategy_scores["ignore"] += 1

    print("Low outlier presence detected.")
    print("Increasing IGNORE preference.\n")

# =========================
# Outlier Strategy Scores
# =========================

print("Outlier Strategy Scores:")
print(outlier_strategy_scores)
```

# Log Transform - First Pass

```python
# =========================
# Log Transform Decision Engine
# =========================

log_transform_scores = {

    "apply": 0,
    "skip": 0
}

# -------------------------
# High Skew -> Apply
# -------------------------

if (
    FEATURE_ENGINEERING_CONFIG[
        "prefer_log_transform_for_skewed_data"
    ]
    and
    len(
        DATASET_METADATA[
            "high_skew_columns"
        ]
    ) > 0
):

    log_transform_scores["apply"] += 2

    print("Highly skewed columns detected.")
    print("Increasing APPLY preference.\n")

# -------------------------
# Low Skew -> Skip
# -------------------------

if (
    FEATURE_ENGINEERING_CONFIG[
        "prefer_skip_for_low_skew_data"
    ]
    and
    len(
        DATASET_METADATA[
            "high_skew_columns"
        ]
    ) == 0
):

    log_transform_scores["skip"] += 1

    print("Low skew dataset detected.")
    print("Increasing SKIP preference.\n")

# -------------------------
# No Positive Columns -> Skip
# -------------------------

if (
    FEATURE_ENGINEERING_CONFIG[
        "require_positive_values_for_log_transform"
    ]
    and
    len(
        DATASET_METADATA[
            "positive_only_columns"
        ]
    ) == 0
):

    log_transform_scores["skip"] += 3

    print("No positive-only columns detected.")
    print("Strongly increasing SKIP preference.\n")

# =========================
# Log Transform Scores
# =========================

print("Log Transform Scores:")
print(log_transform_scores)
```

# Polynomial Handling

```python
# =========================
# Polynomial Decision Engine
# =========================

polynomial_scores = {

    "apply": 0,
    "skip": 0
}

# -------------------------
# Low Feature Count -> Apply
# -------------------------

if (
    FEATURE_ENGINEERING_CONFIG[
        "prefer_polynomial_for_low_feature_count"
    ]
    and
    DATASET_METADATA[
        "feature_count"
    ]
    <
    FEATURE_ENGINEERING_CONFIG[
        "low_feature_count_threshold"
    ]
):

    polynomial_scores["apply"] += 2

    print("Low feature count detected.")
    print("Increasing APPLY preference.\n")

# -------------------------
# High Feature Count -> Skip
# -------------------------

if (
    FEATURE_ENGINEERING_CONFIG[
        "prefer_skip_for_high_feature_count"
    ]
    and
    DATASET_METADATA[
        "feature_count"
    ]
    >
    FEATURE_ENGINEERING_CONFIG[
        "high_feature_count_threshold"
    ]
):

    polynomial_scores["skip"] += 3

    print("High feature count detected.")
    print("Increasing SKIP preference.\n")

# -------------------------
# High Feature-to-Sample Ratio -> Skip
# -------------------------

if (
    DATASET_METADATA[
        "feature_to_sample_ratio"
    ]
    > 0.2
):

    polynomial_scores["skip"] += 2

    print("High feature-to-sample ratio detected.")
    print("Increasing SKIP preference.\n")

# =========================
# Polynomial Feature Scores
# =========================

print("Polynomial Feature Scores:")
print(polynomial_scores)
```

# PCA Handling

```python
# =========================
# Dimensionality Reduction Decision Engine
# =========================

dimensionality_reduction_scores = {

    "pca": 0,
    "skip": 0
}
# high feature count
if (
    FEATURE_ENGINEERING_CONFIG[
        "prefer_pca_for_high_feature_count"
    ]
    and
    DATASET_METADATA[
        "feature_count"
    ]
    >
    FEATURE_ENGINEERING_CONFIG[
        "high_dimensionality_threshold"
    ]
):

    dimensionality_reduction_scores["pca"] += 2

    print("High feature count detected.")
    print("Increasing PCA preference.\n")

#high feature to sample ratio
if (
    FEATURE_ENGINEERING_CONFIG[
        "prefer_pca_for_high_feature_to_sample_ratio"
    ]
    and
    DATASET_METADATA[
        "feature_to_sample_ratio"
    ]
    >
    FEATURE_ENGINEERING_CONFIG[
        "high_feature_to_sample_ratio_threshold"
    ]
):

    dimensionality_reduction_scores["pca"] += 2

    print("High feature-to-sample ratio detected.")
    print("Increasing PCA preference.\n")

#low dimensionality
if (
    FEATURE_ENGINEERING_CONFIG[
        "prefer_skip_for_low_dimensionality"
    ]
    and
    DATASET_METADATA[
        "feature_count"
    ]
    <
    FEATURE_ENGINEERING_CONFIG[
        "low_dimensionality_threshold"
    ]
):

    dimensionality_reduction_scores["skip"] += 2

    print("Low dimensionality detected.")
    print("Increasing SKIP preference.\n")

print("Dimensionality Reduction Scores:")
print(dimensionality_reduction_scores)
```
