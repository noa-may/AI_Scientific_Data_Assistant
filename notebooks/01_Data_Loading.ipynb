# Get File Path - Front End User Window

```python
def select_file():

    while True:

        # Hide tkinter root window
        Tk().withdraw()

        # Open file picker
        file_path = askopenfilename()

        # -------------------------
        # File Successfully Selected
        # -------------------------

        if file_path:

            print("\nFile selected successfully.")
            print(file_path)

            return file_path

        # -------------------------
        # No File Selected
        # -------------------------

        print("\nNo file was selected.")

        print("\nOptions:")
        print("1. Select a file")
        print("2. Close application")

        user_choice = input("\nEnter choice: ")

        # -------------------------
        # Retry File Selection
        # -------------------------

        if user_choice == "1":

            continue

        # -------------------------
        # Exit Application
        # -------------------------

        elif user_choice == "2":

            raise SystemExit(
                "Application closed."
            )

        # -------------------------
        # Invalid Input
        # -------------------------

        else:

            print("\nInvalid choice.")
            print("Please try again.")
```

```python
# -------------------------
# Testing
# -------------------------

file_path = select_file()

print(file_path)
```

# Delimiters

## Define Support Group Files

```python
TEXT_EXTENSIONS = [".csv", ".tsv", ".txt", ".dat"]
EXCEL_EXTENSIONS = [".xlsx", ".xls"]
JSON_EXTENSIONS = [".json"]
PARQUET_EXTENSIONS = [".parquet"]
HDF5_EXTENSIONS = [".h5", ".hdf5"]
```

## Get the File Extension

```python
def get_file_extension(file_path):

    """
    Get lowercase file extension from a file path.
    """

    _, extension = os.path.splitext(file_path)

    return extension.lower()
```

# Detect the Delimiter Dynamically

```python
def detect_delimiter(file_path):

    """
    Automatically detect delimiter for text-based tabular files.
    """

    with open(file_path, "r", encoding="utf-8") as file:

        sample = file.read(4096)

    sniffer = csv.Sniffer()

    try:

        delimiter = sniffer.sniff(sample).delimiter

    except csv.Error:

        print("Could not automatically detect delimiter. Defaulting to comma.")

        delimiter = ","

    return delimiter
```

# Define Loader Functions (per file type)

```python
# -------------------------
# CSV Files, TSV, TXT, DAT
# -------------------------
def load_delimited_text(file_path):

    """
    Load CSV, TSV, TXT, or DAT files using automatic delimiter detection.
    """

    delimiter = detect_delimiter(file_path)

    print(f"Detected delimiter: {repr(delimiter)}")

    df = pd.read_csv(
        file_path,
        sep=delimiter
    )

    return df

# -------------------------
# Excel
# -------------------------
def load_excel(file_path):

    """
    Load Excel spreadsheet files.
    """

    df = pd.read_excel(file_path)

    return df


# -------------------------
# JSON
# -------------------------
def load_json(file_path):

    """
    Load JSON data.
    """

    df = pd.read_json(file_path)

    return df


# -------------------------
# Parquet
# -------------------------
def load_parquet(file_path):

    """
    Load Parquet data.
    """

    df = pd.read_parquet(file_path)

    return df

# -------------------------
# HDF5
# -------------------------
def load_hdf5(file_path):

    """
    Load HDF5 data.

    Note: This assumes the HDF5 file has a pandas-readable table structure.
    Some scientific HDF5 files may need custom parsing later.
    """

    df = pd.read_hdf(file_path)

    return df
```

# Load Data into Pandas Dataframe

```python
FILE_LOADERS = {
    ".csv": load_delimited_text,
    ".tsv": load_delimited_text,
    ".txt": load_delimited_text,
    ".dat": load_delimited_text,

    ".xlsx": load_excel,
    ".xls": load_excel,

    ".json": load_json,

    ".parquet": load_parquet,

    ".h5": load_hdf5,
    ".hdf5": load_hdf5
}
```

```python
def load_data(file_path):

    """
    Load structured data from multiple file formats into a pandas DataFrame.
    """

    if not file_path:

        raise ValueError("No file path was provided.")

    extension = get_file_extension(file_path)

    if extension not in FILE_LOADERS:

        raise ValueError(
            f"Unsupported file format: {extension}"
        )

    loader_function = FILE_LOADERS[extension]

    df = loader_function(file_path)

    print("\nDataset Loaded Successfully")
    print("-" * 40)
    print(f"File type: {extension}")
    print(f"Rows: {df.shape[0]}")
    print(f"Columns: {df.shape[1]}")

    print("\nColumn Names:")
    print(df.columns.tolist())

    return df
```
